```python
# Task 1: Read a Text File and Count Lines, Words, and Characters
import logging

# Setup logging for error handling
logging.basicConfig(filename='error_log.txt', level=logging.ERROR)

def count_file_contents(input_file, output_file):
    try:
        with open(input_file, 'r') as file:
            content = file.read()

        lines = content.splitlines()
        words = content.split()
        characters = len(content)

        with open(output_file, 'w') as result_file:
            result_file.write(f"Lines: {len(lines)}\n")
            result_file.write(f"Words: {len(words)}\n")
            result_file.write(f"Characters: {characters}\n")

        print(f"Results written to {output_file}")

    except FileNotFoundError:
        print("The specified file was not found.")
        logging.error("FileNotFoundError: The specified file was not found.")
    except Exception as e:
        print(f"An error occurred: {e}")
        logging.error(f"Error: {e}")

# File paths
text_file_path = r"C:\Users\Administrator\Desktop\students_data.csv"
output_text_path = r"C:\Users\Administrator\Desktop\students_data.csv"

# Example usage of Task 1
print("Task 1: Count lines, words, and characters")
count_file_contents(text_file_path, output_text_path)
```

```python
import csv
import logging

# Setup logging for error handling
logging.basicConfig(filename='error_log.txt', level=logging.ERROR)

def calculate_average_expenses(input_csv, output_csv):
    try:
        student_expenses = {}

        with open(input_csv, newline='') as csvfile:
            reader = csv.DictReader(csvfile)

            for row in reader:
                try:
                    # Retrieve necessary columns
                    student = row['Course']
                    semester_expense = float(row['Expense_Semester'])
                    accommodation_expense = float(row['Expense_Accommodation'])

                    # Calculate total expenses
                    total_expense = semester_expense + accommodation_expense
                    student_expenses[student] = total_expense

                except ValueError:
                    print(f"Invalid numeric value encountered for {student}.")
                    logging.error(f"ValueError: Invalid numeric value encountered for {student}.")
                    continue

        # Write the results to a new CSV file
        with open(output_csv, 'w', newline='') as csvfile:
            fieldnames = ['Course', 'Total Expenses']
            writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
            writer.writeheader()
            for student, total_expense in student_expenses.items():
                writer.writerow({'Course': student, 'Total Expenses': total_expense})

        print(f"Average expenses written to {output_csv}")

    except FileNotFoundError:
        print("The specified CSV file was not found.")
        logging.error("FileNotFoundError: The specified CSV file was not found.")
    except KeyError as e:
        print(f"Missing expected column in CSV file: {e}")
        logging.error(f"KeyError: Missing expected column {e} in CSV file.")
    except Exception as e:
        print(f"An error occurred: {e}")
        logging.error(f"Error: {e}")

# File paths
csv_file_path = r"C:\Users\Administrator\Desktop\students-data.csv"
output_csv_path = r"C:\Users\Administrator\Desktop\average_expenses.csv"

# Example usage of Task 2
print("\nTask 2: Calculate average expenses")
calculate_average_expenses(csv_file_path, output_csv_path)
```
